````
from google.colab import drive
import pandas as pd
import os
from collections import defaultdict

# Mount Google Drive
drive.mount('/content/drive')

# อ่านข้อมูลทั้งหมดจากโฟลเดอร์ย่อยใน base_path
def read_all_data(base_path):
    all_data = []
    for month in os.listdir(base_path):
        month_path = os.path.join(base_path, month)
        if not os.path.isdir(month_path):
            continue
        for file in os.listdir(month_path):
            if 'stop-and-search' not in file:
                continue
            stopandsearchpath = os.path.join(month_path, file)
            try:
                df = pd.read_csv(stopandsearchpath)
                # ตัดชื่อไฟล์เพื่อเอาแค่ชื่อเมือง เช่น "avon-and-somerset"
                city = file.replace('-stop-and-search.csv', '').split('-', 2)[-1].strip()
                df['Police Force'] = city
                all_data.append(df)
            except Exception as e:
                print(f"Error reading {stopandsearchpath}: {e}")
    return all_data

# นับจำนวนคนเชื้อสายเอเชีย
def count_asian_people(all_dfs):
    asian_counts = defaultdict(int)
    for df in all_dfs:
        asian_df = df[df['Self-defined ethnicity'].str.contains('Asian', na=False)]
        if not asian_df.empty:
            city = asian_df['Police Force'].iloc[0]
            asian_counts[city] += len(asian_df)
    return dict(asian_counts)

# === Main execution ===
base_path = '/content/drive/MyDrive/Python&DataViz2025'

# โหลดข้อมูล
dataframes = read_all_data(base_path)

# คำนวณจำนวนคนเชื้อสายเอเชีย
counts = count_asian_people(dataframes)

# แสดงผล
if counts:
    most_city = max(counts, key=counts.get)
    fewest_city = min(counts, key=counts.get)

    print(f"เมืองที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด คือเมือง {fewest_city} จำนวน {counts[fewest_city]} คน")
    print(f"เมืองที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia มากที่สุด คือเมือง {most_city} จำนวน {counts[most_city]} คน")
else:
    print("ไม่พบข้อมูลที่เกี่ยวข้องกับผู้มีเชื้อสายเอเชีย")
````

````
from google.colab import drive
import pandas as pd
import os
from collections import defaultdict

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)

# Function to read all data for a given month and count Asian people
def count_asian_in_month(month_path):
    asian_count = 0
    if not os.path.isdir(month_path):
        return 0

    for file in os.listdir(month_path):
        if 'stop-and-search' in file:
            filepath = os.path.join(month_path, file)
            try:
                df = pd.read_csv(filepath)
                asian_df = df[df['Self-defined ethnicity'].str.contains('Asian', na=False)]
                asian_count += len(asian_df)
            except Exception as e:
                print(f"Error reading {filepath}: {e}")
    return asian_count

# === Main execution ===
base_path = '/content/drive/MyDrive/Python&DataViz2025'
monthly_asian_counts = {}

# Iterate through each month folder
for month in os.listdir(base_path):
    month_path = os.path.join(base_path, month)
    if os.path.isdir(month_path):
      monthly_asian_counts[month] = count_asian_in_month(month_path)

# Find the month with the most and fewest Asian stop and searches
if monthly_asian_counts:
    most_month = max(monthly_asian_counts, key=monthly_asian_counts.get)
    fewest_month = min(monthly_asian_counts, key=monthly_asian_counts.get)

    print("Total stop and searches for Asian people per month:\n")
    for month in sorted(monthly_asian_counts):
        print(f"{month}: {monthly_asian_counts[month]}")

    print("\n====== Summary ======")
    print(f"Month with the most stop and searches for Asian people: {most_month} ({monthly_asian_counts[most_month]})")
    print(f"Month with the fewest stop and searches for Asian people: {fewest_month} ({monthly_asian_counts[fewest_month]})")
else:
    print("No data found or processed.")
````

````
# นำเข้าไลบรารีที่จำเป็น
from google.colab import drive # นำเข้าโมดูล drive จาก google.colab เพื่อเชื่อมต่อ Google Drive
import pandas as pd # นำเข้าไลบรารี pandas สำหรับทำงานกับข้อมูลแบบ DataFrame
import os # นำเข้าโมดูล os สำหรับการทำงานกับระบบไฟล์และไดเร็กทอรี
from collections import defaultdict # นำเข้า defaultdict จาก collections ซึ่งเป็น subclass ของ dict ที่กำหนดค่าเริ่มต้นสำหรับ key ที่ไม่มี

# เชื่อมต่อ Google Drive
drive.mount('/content/drive', force_remount=True) # เชื่อมต่อ Google Drive เข้ากับ Colab โดย mount ไปที่ '/content/drive' และใช้ force_remount=True เพื่อบังคับเชื่อมต่อใหม่หากเชื่อมต่ออยู่แล้ว

# กำหนดฟังก์ชันเพื่ออ่านข้อมูล stop-and-search สำหรับเดือนที่กำหนดและนับจำนวนคนเชื้อสายเอเชีย
def count_asian_in_month(month_path): # กำหนดฟังก์ชันชื่อ count_asian_in_month ที่รับพารามิเตอร์ month_path (เส้นทางของโฟลเดอร์เดือน)
    asian_count = 0 # กำหนดตัวแปร asian_count เริ่มต้นที่ 0 เพื่อนับจำนวนคนเชื้อสายเอเชีย
    if not os.path.isdir(month_path): # ตรวจสอบว่า month_path เป็นไดเร็กทอรีหรือไม่
        return 0 # ถ้าไม่ใช่ไดเร็กทอรี ให้คืนค่า 0

    for file in os.listdir(month_path): # วนลูปผ่านไฟล์ทั้งหมดในไดเร็กทอรี month_path
        if 'stop-and-search' in file: # ตรวจสอบว่าชื่อไฟล์มีคำว่า 'stop-and-search' อยู่หรือไม่
            filepath = os.path.join(month_path, file) # สร้าง filepath ที่สมบูรณ์สำหรับไฟล์นั้น
            try:
                df = pd.read_csv(filepath) # อ่านไฟล์ CSV เข้ามาเก็บใน DataFrame ชื่อ df
                # กรอง DataFrame เพื่อเลือกเฉพาะแถวที่คอลัมน์ 'Self-defined ethnicity' มีคำว่า 'Asian'
                asian_df = df[df['Self-defined ethnicity'].str.contains('Asian', na=False)] # กรอง df โดยเลือกแถวที่คอลัมน์ 'Self-defined ethnicity' มีคำว่า 'Asian' (ไม่รวมค่าว่างในคอลัมน์นี้) และเก็บผลลัพธ์ใน asian_df
                asian_count += len(asian_df) # เพิ่มจำนวนแถวใน asian_df (จำนวนคนเชื้อสายเอเชียในไฟล์นี้) เข้าไปใน asian_count
            except Exception as e: # ดักจับข้อผิดพลาดที่อาจเกิดขึ้นระหว่างอ่านไฟล์
                print(f"Error reading {filepath}: {e}") # พิมพ์ข้อความแจ้งเตือนข้อผิดพลาดพร้อมชื่อไฟล์
    return asian_count # คืนค่า asian_count ซึ่งเป็นจำนวนคนเชื้อสายเอเชียในเดือนนั้นๆ

# ส่วนการทำงานหลักของโปรแกรม
base_path = '/content/drive/MyDrive/Python&DataViz2025' # กำหนด base_path ไปยังโฟลเดอร์หลักที่เก็บข้อมูล
monthly_asian_counts = {} # สร้าง dictionary ชื่อ monthly_asian_counts เพื่อเก็บจำนวนคนเชื้อสายเอเชียในแต่ละเดือน

# วนลูปผ่านแต่ละโฟลเดอร์เดือนใน base_path
for month in os.listdir(base_path): # วนลูปผ่านชื่อไฟล์และโฟลเดอร์ทั้งหมดใน base_path
    month_path = os.path.join(base_path, month) # สร้าง month_path ที่สมบูรณ์สำหรับแต่ละรายการ
    if os.path.isdir(month_path): # ตรวจสอบว่า month_path เป็นไดเร็กทอรีหรือไม่ (เพื่อข้ามไฟล์ที่ไม่ใช่โฟลเดอร์เดือน)
      monthly_asian_counts[month] = count_asian_in_month(month_path) # เรียกใช้ฟังก์ชัน count_asian_in_month เพื่อนับจำนวนคนเอเชียในเดือนนั้นๆ และเก็บผลลัพธ์ลงใน monthly_asian_counts โดยใช้ชื่อเดือนเป็น key

# หาเดือนที่มีจำนวนคนเชื้อสายเอเชียถูกเรียกค้นมากที่สุดและน้อยที่สุด
if monthly_asian_counts: # ตรวจสอบว่า monthly_asian_counts ไม่ว่างเปล่า
    most_month = max(monthly_asian_counts, key=monthly_asian_counts.get) # หา key (ชื่อเดือน) ที่มี value (จำนวนคนเอเชีย) มากที่สุดใน monthly_asian_counts
    fewest_month = min(monthly_asian_counts, key=monthly_asian_counts.get) # หา key (ชื่อเดือน) ที่มี value (จำนวนคนเอเชีย) น้อยที่สุดใน monthly_asian_counts

    # แสดงผลรวมจำนวนคนเชื้อสายเอเชียที่ถูกเรียกค้นในแต่ละเดือน
    print("Total stop and searches for Asian people per month:\n") # พิมพ์หัวข้อ
    for month in sorted(monthly_asian_counts): # วนลูปผ่านชื่อเดือนใน monthly_asian_counts โดยเรียงตามลำดับ
        print(f"{month}: {monthly_asian_counts[month]}") # พิมพ์ชื่อเดือนและจำนวนคนเชื้อสายเอเชียในเดือนนั้นๆ

    # แสดงสรุปผล
    print("\n====== Summary ======") # พิมพ์หัวข้อสรุป
    print(f"Month with the most stop and searches for Asian people: {most_month} ({monthly_asian_counts[most_month]})") # พิมพ์เดือนที่มีจำนวนคนเอเชียถูกเรียกค้นมากที่สุดและจำนวน
    print(f"Month with the fewest stop and searches for Asian people: {fewest_month} ({monthly_asian_counts[fewest_month]})") # พิมพ์เดือนที่มีจำนวนคนเอเชียถูกเรียกค้นน้อยที่สุดและจำนวน
else:
    print("No data found or processed.") # ถ้า monthly_asian_counts ว่างเปล่า ให้พิมพ์ข้อความนี้
````

````
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import os

# ฟังก์ชันอ่านไฟล์ทั้งหมด เก็บข้อมูลลง list ที่เพิ่มคอลัมน์ชื่อเมืองเข้าไป
def read_all_data(base_path):
    all_dfs = []  # สร้าง list เปล่าเพื่อเอาไว้เก็บไฟล์ทั้งหมด
    months = os.listdir(base_path)  # สร้างตัวแปรเพื่อเก็บโฟลเดอร์ดือนทั้งหมด

    for month in months:
        month_path = os.path.join(base_path, month) # เชื่อม basepath กับชื่อเดือนเป็นเส้นทางของแต่ละเดือน
        if not os.path.isdir(month_path):
            continue
        files = os.listdir(month_path) #สร้างตัวแปรเพื่อเก็บชื่อไฟล์ทั้งหมดที่อยู่ในเดือนนั้น
        for file in files:
            if 'stop-and-search' in file: #ดำเนินการเฉพาะไฟล์ 'stop-and-search' ทุกเมือง
                filepath = os.path.join(month_path, file) # สร้างไฟล์พาทที่มีไฟล์เฉพาะ stop-and-search
                df = pd.read_csv(filepath) # สร้าตัวแปรมารับตาราง
                city = file.replace('-stop-and-search.csv', '').split('-', 2)[-1] # เลือกมาแค่ชื่อเมือง
                df['Police Force'] = city # นำชื่อเมืองมาสร้างคอลัมน์ใหม่ใน df
                all_dfs.append(df) #เพิ่ม df ใน all_dfs

    return all_dfs #คืนค่าตารางทั้งหมดที่้พิ่มเข้ามาเรือยๆ

# ฟังก์ชันนับจำนวนคนเชื้อสายเอเชีย รวมทุกไฟล์
def count_asian_people(all_dfs):
    asian_counts = {}  # สร้าง dictionary ว่าง เพื่อเก็บจำนวนคนเชื้อสายเอเชียของแต่ละเมือง

    for df in all_dfs:  # วนลูปดูข้อมูลในแต่ละ DataFrame ที่เก็บมาใน list
        # Filter for rows where 'Self-defined ethnicity' contains 'Asian'
        asian_df = df[df['Self-defined ethnicity'].str.contains('Asian', na=False)]
        if not asian_df.empty:
            city = asian_df['Police Force'].iloc[0]

            if city in asian_counts:
                asian_counts[city] += len(asian_df)
            else:
                asian_counts[city] = len(asian_df)

    return asian_counts  # คืนค่า dictionary ที่เก็บจำนวนคนเชื้อสายเอเชียของแต่ละเมือง

# ส่วนหลักของโปรแกรม
base_path = '/content/drive/MyDrive/Python&DataViz2025'

# อ่านข้อมูลจากไฟล์ทั้งหมด
dataframes = read_all_data(base_path)

# นับคนเชื้อสายเอเชียรวมทั้ง 5 เดือน
counts = count_asian_people(dataframes)

# หามากที่สุดและน้อยที่สุด
if counts:
    most_city = max(counts, key=counts.get)
    fewest_city = min(counts, key=counts.get)

    # แสดงผลทุกเมือง
    print("รวมจำนวนการเรียกค้นผู้มีเชื้อสายเอเชียในแต่ละเมือง (รวม 5 เดือน):\n")
    for city in sorted(counts, key=counts.get, reverse=True):
        print(f"{city}: {counts[city]}")

    print("\n====== สรุป ======")
    print(f"เมืองที่มีการเรียกค้นคนเชื้อสายเอเชียมากที่สุด: {most_city} ({counts[most_city]})")
    print(f"เมืองที่มีการเรียกค้นคนเชื้อสายเอเชียน้อยที่สุด: {fewest_city} ({counts[fewest_city]})")
else:
    print("No data found or processed.")
````

### คำอธิบายฟังก์ชัน `read_all_data(base_path)`
ฟังก์ชัน `read_all_data` มีหน้าที่ในการอ่านข้อมูลไฟล์ CSV ที่เกี่ยวข้องกับ "stop-and-search" จากโฟลเดอร์ต่างๆ ที่จัดเก็บตามเดือนในปี 2025 โดยมีขั้นตอนการทำงานดังนี้:
1. **เริ่มต้นด้วย list ว่าง:** สร้าง list ชื่อ `all_data` เพื่อใช้เก็บ DataFrame ของแต่ละไฟล์
2. **วนลูปในโฟลเดอร์เดือน:** เข้าถึงโฟลเดอร์หลัก (`base_path`) และวนลูปเพื่อเข้าถึงโฟลเดอร์ของแต่ละเดือน (เช่น '2025-01', '2025-02', ...)
3. **วนลูปในไฟล์:** ในแต่ละโฟลเดอร์เดือน จะวนลูปเพื่อเข้าถึงไฟล์ทั้งหมด
4. **กรองไฟล์ "stop-and-search":** ตรวจสอบว่าชื่อไฟล์มีคำว่า "stop-and-search" หรือไม่ เพื่อให้แน่ใจว่าเราจะประมวลผลเฉพาะไฟล์ที่ต้องการ
5. **อ่านไฟล์ CSV:** หากไฟล์นั้นเป็นไฟล์ "stop-and-search" จะใช้ `pandas.read_csv()` เพื่ออ่านข้อมูลในไฟล์นั้นๆ
6. **เพิ่มคอลัมน์ "Police Force":**
   - **ดึงชื่อเมืองจากชื่อไฟล์:** ทำการตัดคำว่า "-stop-and-search.csv" ออกจากชื่อไฟล์ และใช้ `.split('-', 2)[-1].strip()` เพื่อให้ได้ชื่อเมืองที่สะอาด (เช่น "avon-and-somerset")
   - **สร้างคอลัมน์ใหม่:** เพิ่มคอลัมน์ใหม่ชื่อ "Police Force" ใน DataFrame และใส่ชื่อเมืองที่ดึงมาได้
7. **รวม DataFrame:** เพิ่ม DataFrame ที่ผ่านการประมวลผลแล้วลงใน list `all_data`
8. **คืนค่า:** เมื่อวนลูปครบทุกไฟล์และทุกโฟลเดอร์แล้ว ฟังก์ชันจะคืนค่า `all_data` ซึ่งเป็น list ที่รวบรวม DataFrame ของข้อมูล "stop-and-search" จากทุกเมืองและทุกเดือน

### คำอธิบายฟังก์ชัน `count_asian_people(all_dfs)`
ฟังก์ชัน `count_asian_people` ถูกออกแบบมาเพื่อคำนวณจำนวนการเรียกค้นบุคคลเชื้อสายเอเชียในแต่ละเมือง โดยรับ list ของ DataFrames (`all_dfs`) เป็น input และมีขั้นตอนการทำงานดังนี้:

1. **สร้าง defaultdict:** `asian_counts = defaultdict(int)` สร้าง `defaultdict` ซึ่งเป็น Dictionary ประเภทพิเศษที่กำหนดค่าเริ่มต้นเป็น 0 สำหรับคีย์ใหม่ ซึ่งมีประโยชน์ในการนับจำนวนโดยไม่ต้องตรวจสอบว่าคีย์มีอยู่แล้วหรือไม่
2. **วนลูปในแต่ละ DataFrame:** ฟังก์ชันจะวนลูปผ่าน DataFrame แต่ละอันใน `all_dfs`
3. **กรองข้อมูลคนเชื้อสายเอเชีย:** `asian_df = df[df['Self-defined ethnicity'].str.contains('Asian', na=False)]` ใช้ `.str.contains('Asian', na=False)` เพื่อกรองแถวที่คอลัมน์ `Self-defined ethnicity` มีคำว่า "Asian" อยู่ และเก็บผลลัพธ์ใน `asian_df`
4. **ตรวจสอบข้อมูล:** `if not asian_df.empty:` ตรวจสอบว่า `asian_df` มีข้อมูลหรือไม่ (นั่นคือ มีคนเอเชียใน DataFrame นั้นๆ หรือไม่)
5. **ดึงชื่อเมือง:** `city = asian_df['Police Force'].iloc[0]` หากมีข้อมูล จะดึงชื่อเมืองจากคอลัมน์ "Police Force" ของแถวแรกใน `asian_df`
6. **นับจำนวน:** `asian_counts[city] += len(asian_df)` เพิ่มจำนวนคนเอเชีย (ขนาดของ `asian_df`) เข้าไปใน `asian_counts` โดยใช้ชื่อเมืองเป็นคีย์
7. **คืนค่า:** `return dict(asian_counts)` แปลง `defaultdict` กลับเป็น dictionary ธรรมดาและคืนค่า ซึ่งเก็บข้อมูลจำนวนคนเชื้อสายเอเชียในแต่ละเมือง

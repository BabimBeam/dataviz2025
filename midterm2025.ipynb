```python
# --- เริ่มการสืบสวน ---
print(">> ขั้นตอนที่ 1: ค้นหารายงานจากที่เกิดเหตุ")

# ค้นหารายงานคดีฆาตกรรม (murder) ที่เมือง Pandas City ในวันที่ 15 มกราคม 2018 (20180115)
target_crime_report = crime_df[
    (crime_df['date'] == 20180115) &
    (crime_df['type'] == 'murder') &
    (crime_df['city'] == 'Pandas City')
]

# ดึงรายละเอียด (description) จากรายงานที่พบ
crime_description = target_crime_report['description'].iloc[0]

print("✅ พบรายงานคดี! รายละเอียดจากที่เกิดเหตุระบุว่า:")
print(f'"{crime_description}"')
print("\nจากข้อมูลนี้ ผมพบว่ามีพยาน 2 คน คือ Annabel Miller (ID: 14887) และ Morty Schapiro (ID: 16371)")
# 1. กำหนดคำสำคัญ (Keywords) ที่เกี่ยวข้องกับคดีฆาตกรรม
# ผมเลือกคำว่า 'gunshot' (เสียงปืน) และ 'hired' (ถูกจ้าง) เพราะเป็นคำที่บ่งชี้ถึงอาชญากรรมร้ายแรง
keywords = ['gunshot', 'hired']

# 2. สร้างเงื่อนไขการค้นหา (Regular Expression) เพื่อหาคำใดคำหนึ่งในรายการ
# '|' หมายถึง "หรือ" (OR)
search_pattern = '|'.join(keywords)

# 3. ค้นหาในคอลัมน์ 'transcript' ของ interview_df
# เราจะกรองเฉพาะแถวที่ในคำให้การมีคำสำคัญที่เรากำหนดไว้
# `na=False` เพื่อให้ไม่เกิด error หากมีช่องว่าง
relevant_interviews = interview_df[interview_df['transcript'].str.contains(search_pattern, na=False)]

# 4. แสดงผลลัพธ์
print("✅ พบคำให้การที่คาดว่าเกี่ยวข้องกับคดีฆาตกรรม:")
print("--------------------------------------------------")

# วนลูปเพื่อแสดงผลลัพธ์ที่หาเจอทีละรายการ
for index, row in relevant_interviews.iterrows():
    print(f"Person ID: {row['person_id']}")
    print(f"Transcript: {row['transcript']}")
    print("--------------------------------------------------")

```

```python
print("\n>> ขั้นตอนที่ 2: ตามหาฆาตกรจากเบาะแส")

# 1. ค้นหาสมาชิกยิมที่ตรงตามเงื่อนไข (ID '48Z' และสถานะ 'gold')
gym_suspects = member_df[
    (member_df['id'].str.startswith('48Z')) &
    (member_df['membership_status'] == 'gold')
]

# 2. นำผลที่ได้ไปเชื่อมกับข้อมูลบุคคล (person_df) และใบขับขี่ (drivers_license_df)
person_suspects = pd.merge(gym_suspects, person_df, left_on='person_id', right_on='id')
license_suspects = pd.merge(person_suspects, drivers_license_df, left_on='license_id', right_on='id')

# 3. กรองขั้นสุดท้ายด้วยข้อมูลป้ายทะเบียนรถ
murderer_info = license_suspects[license_suspects['plate_number'].str.contains("H42W")]

print("✅ พบตัวฆาตกรแล้ว!")

# --- แสดงผลลัพธ์ของฆาตกร ---
murderer_name = murderer_info['name_x'].values[0]
murderer_car = f"{murderer_info['car_make'].values[0]} {murderer_info['car_model'].values[0]}"


print("\n=== 🔪 ฆาตกร  ===")
print(f"ชื่อ: {murderer_name}")

```

```python
# ==============================================================================
#   ขั้นตอนที่ 3: ตามหาผู้บงการ (The Mastermind)
# ==============================================================================
print("\n>> ขั้นตอนที่ 3: กำลังตามหาผู้บงการจากคำให้การของฆาตกร...")

# จากการสอบสวน ฆาตกร (Jeremy Bowers) ให้การว่า:
# "ถูกจ้างโดยผู้หญิงรวย, สูงประมาณ 65-67 นิ้ว, ผมสีแดง, ขับรถ Tesla Model S
# และไปคอนเสิร์ต SQL Symphony 3 ครั้งในเดือนธันวาคม 2017"

# --- 3.1: ค้นหาผู้ที่ตรงตามโปรไฟล์กิจกรรม ---
print("[3.1] ค้นหาผู้ที่ไปคอนเสิร์ต 'SQL Symphony' 3 ครั้งในเดือนธันวาคม 2017...")
event_attendees = facebook_df[
    (facebook_df['event_name'] == 'SQL Symphony Concert') &
    (facebook_df['date'].between(20171201, 20171231))
]
attendance_counts = event_attendees['person_id'].value_counts()
potential_suspect_ids = attendance_counts[attendance_counts == 3].index.tolist()

print(f"✅ พบผู้ที่ไปคอนเสิร์ต 3 ครั้ง ทั้งหมด {len(potential_suspect_ids)} คน")

# --- 3.2: กรองผู้ต้องสงสัยด้วยข้อมูลทางกายภาพและรถยนต์ ---
print("[3.2] กรองผู้ต้องสงสัยด้วยโปรไฟล์ 'ผู้หญิงผมแดงขับ Tesla'...")
potential_suspects_details = person_df[person_df['id'].isin(potential_suspect_ids)]
suspects_with_license = pd.merge(potential_suspects_details, drivers_license_df, left_on='license_id', right_on='id')

mastermind_info = suspects_with_license[
    (suspects_with_license['gender'] == 'female') &
    (suspects_with_license['height'].between(65, 67)) &
    (suspects_with_license['hair_color'] == 'red') &
    (suspects_with_license['car_make'] == 'Tesla') &
    (suspects_with_license['car_model'] == 'Model S')
]

# --- สรุปข้อมูลผู้บงการ ---
mastermind_name = mastermind_info['name'].iloc[0]
mastermind_car = f"{mastermind_info['car_make'].iloc[0]} {mastermind_info['car_model'].iloc[0]}"

print("✅ พบตัวผู้บงการแล้ว!")
print(f"ชื่อ: {mastermind_name}")


# ==============================================================================
#   สรุปผลการสืบสวน
# ==============================================================================
print("\n--- 📜 สรุปผลการสืบสวนฉบับสมบูรณ์ ---")
print(f"**ฆาตกร:** \t\t{murderer_name}")
print("-" * 40)
print(f"**ผู้บงการ:** \t\t{mastermind_name}")
```
